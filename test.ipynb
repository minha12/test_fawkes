## Spliting train/test set

```python
from imutils import paths
```

```python
image_paths = list(paths.list_images('imm/') )
train_imgs = [x for x in image_paths if x.split('-')[1][0] != '6']
train_imgs.sort()
test_imgs = [x for x in image_paths if x.split('-')[1][0] == '6']
test_imgs.sort()
test_imgs
```

```python
import os
import shutil
src = 'imm/'
dst = 'imm/train/'
def move_files(img_paths, dst):
    if not os.path.exists(dst):
        os.makedirs(dst)

    for f in img_paths:
        shutil.move(f, dst)
```
